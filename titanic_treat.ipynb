```python
import os
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import plotly as plt
```

```python
FOLDER_TRAIN =  os.path.join('C://Users', 'Hugo Martins',
                       'downloads', 'train.csv')
FOLDER_TEST =  os.path.join('C://Users', 'Hugo Martins',
                       'downloads', 'test.csv')
```

```python
train_data = pd.read_csv(FOLDER_TRAIN)
train_data.head()
train_data.shape
```

```python
test_data = pd.read_csv(FOLDER_TEST)
test_data
```

```python
male = train_data.loc[train_data.Sex == 'male']["Survived"]
female = train_data.loc[train_data.Sex == 'female']["Survived"]

rate_male = sum(male)/len(male)
rate_female = sum(female)/len(female)

rate_female, rate_male
```

```python
y = train_data['Survived']
columns = ['Pclass', 'Parch', 'Sex', 'SibSp']
x = pd.get_dummies(train_data[columns])
x_test = pd.get_dummies(test_data[columns])

x
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(x,y)
predictions = model.predict(x_test)

output = pd.DataFrame({'PassengerID': test_data.PassengerId, 'Survived': predictions,
                       'Sex': test_data.Sex})
# output.to_csv('submission.csv', index=False)

```

```python
survived = sum(output.Survived)/len(output.Survived)
male = output[output.Sex == 'male']['Survived']
female = output[output.Sex == 'female']['Survived']
rate_male_out = sum(male)/len(male)
rate_female_out = sum(female)/len(female)
survived, rate_male, rate_female
```
